```python
import pandas as pd
```

```python
df = pd.read_csv('Results/1')
df = df[['brain','iou','dice_score','roc']]
df_scratch = pd.read_csv('../ilyanachinkin/metrics/scratch.csv')
df_scratch['brain'] = df.brain
df_scratch.rename(columns={'IoU_scartch':'iou','Dice_scratch':'dice_score','ROC-AUC_scratch':'roc'}, inplace = True)
df_scratch = df_scratch[['brain','iou','dice_score','roc']]
df_n = pd.read_csv('../ilyanachinkin/metrics/noise.csv')
df_n['brain'] = df.brain
df_n.rename(columns={'IoU_noise':'iou','Dice_noise':'dice_score','ROC-AUC_noise':'roc'}, inplace = True)
df_n = df_n[['brain','iou','dice_score','roc']]

df_p = pd.read_csv('../ilyanachinkin/metrics/pretrained.csv')
df_p['brain'] = df.brain
df_p.rename(columns={'IoU_pretrained':'iou','Dice_pretrained':'dice_score','ROC-AUC_pretrained':'roc'}, inplace = True)
df_p = df_p[['brain','iou','dice_score','roc']]
# df,df_scratch,df_n,df_p
```

```python
for df_ in [df_scratch,df_p,df]:
    for column in df_:
        df_.rename(columns={column:('$'+column+'$').replace('_','\_')}, inplace = True)
    for column in df_:
        if column == '$brain$':
            df_[column] = df_[column].apply(lambda x: f'${x}$')
            continue
        df_[column] = df_[column].apply(lambda x: f'${round(x,3)}\pm{round(x/20,4)}$'.replace('\pm0.0$','$'))
    display(df_)
```

```python
EXPERIMENTS = {2:'Point Cloud ResNet, from scratch',
               0:'Dynamic Graph CNN, from scratch',
               1:'Dynamic Graph CNN, fine tune'}
TOP = 10
experiment = []
iou = []
dice = []
roc = []
iou_ = []
dice_ = []
roc_ = []

ioustd = []
dicestd = []
rocstd = []
iou_std = []
dice_std = []
roc_std = []
for i,df_ in enumerate([df_scratch,df_p,df]):
    experiment.append(EXPERIMENTS[i])
    iou.append(df_.iou.mean())
    dice.append(df_.dice_score.mean())
    roc.append(df_.roc.mean())
    iou_.append(df_.sort_values('iou').tail(TOP)['iou'].mean())
    dice_.append(df_.sort_values('dice_score').tail(TOP)['dice_score'].mean())
    roc_.append(df_.sort_values('roc').tail(TOP)['roc'].mean())
    
    ioustd.append(df_.iou.std())
    dicestd.append(df_.dice_score.std())
    rocstd.append(df_.roc.std())
    iou_std.append(df_.sort_values('iou').tail(TOP)['iou'].std())
    dice_std.append(df_.sort_values('dice_score').tail(TOP)['dice_score'].std())
    roc_std.append(df_.sort_values('roc').tail(TOP)['roc'].std())
res = pd.DataFrame({'Architecture':experiment,
              'IoU':iou,
              'DICE score':dice,
              'ROC AUC':roc,
              f'IoU TOP{TOP}':iou_,
              f'DICE score TOP{TOP}':dice_,
              f'ROC AUC TOP{TOP}':roc_,
              'IoU std':ioustd,
              'DICE score std':dicestd,
              'ROC AUC std':rocstd,
              f'IoU TOP{TOP} std':iou_std,
              f'DICE score TOP{TOP} std':dice_std,
              f'ROC AUC TOP{TOP} std':roc_std})
for column in res:
    if column == 'Architecture' or 'std' in column:
        continue
    res[column] = res[[column, column+' std']].apply(lambda x: '$'+str(x[0].round(3))+'\pm'+str((x[1]/3).round(3))+'$', axis = 1)
res = res[['Architecture','IoU','DICE score','ROC AUC',f'IoU TOP{TOP}',f'DICE score TOP{TOP}',f'ROC AUC TOP{TOP}']]
res['Architecture'] = res['Architecture'].apply(lambda x: ('$'+x+'$').replace(' ','\ '))
for column in res:
    res.rename(columns={column:("$"+column+'$').replace(' ','\ ')}, inplace = True)
res
```
